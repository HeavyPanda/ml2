###  Ecercise 1
####  i)

$$
\min_{w} {E(w) = \sum_i {\left|\vec{x}_i-\sum_j{w_{ij} \vec{x}_j}\right|^2}}
$$
$$
s.t. \sum_j{w_{ij}} = 1
$$
$$
\equiv
$$
$$
\min_{w} {\sum_i {\left|\alpha x_i-\sum_j{w_{ij} \alpha x_j}\right|^2}}
$$
$$
s.t. \sum_j{w_{ij}} = 1
$$
$$
\equiv
$$
$$
\min_{w} {\sum_i {\left|\alpha x_i-\alpha \sum_j{w_{ij} x_j}\right|^2}}
$$
$$
s.t. \sum_j{w_{ij}} = 1
$$
$$
\equiv
$$
$$
\min_{w} {\sum_i {\alpha^2 \left| x_i - \sum_j{w_{ij} x_j}\right|^2}}
$$
$$
s.t. \sum_j{w_{ij}} = 1
$$
$$
\equiv
$$
$$
\min_{w} \alpha^2 {\sum_i {\left| x_i - \sum_j{w_{ij} x_j}\right|^2}}
$$
$$
s.t. \sum_j{w_{ij}} = 1
$$

Since the multiplication by $\alpha^2 \in \mathbb{R}^+ \setminus \{0\}$ doesn't change the minima with respect to $w$, the minimizaiton problem remains the same
